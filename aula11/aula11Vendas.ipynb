```python
# Dicionário de vendas do mês de Agosto
vendas_agosto = {
    "04/08/2025": 2350,
    "05/08/2025": 1980,
    "06/08/2025": 2540,
    "07/08/2025": 2100,
    "08/08/2025": 3050,
    "11/08/2025": 2650,
    "12/08/2025": 2180,
    "13/08/2025": 2760,
    "14/08/2025": 2920,
    "15/08/2025": 3100,
    "18/08/2025": 2300,
    "19/08/2025": 2430,
    "20/08/2025": 2210,
    "21/08/2025": 2790,
    "22/08/2025": 2970,
    "25/08/2025": 3100,
    "26/08/2025": 2880,
    "27/08/2025": 2600,
    "28/08/2025": 2500,
    "29/08/2025": 3400
}

valores_vendas = list(vendas_agosto.values())
lista_datas = list(vendas_agosto.keys())

# Cálculos de Totais e Média Mensal
vendas_mes = sum(vendas_agosto.values())
media_diaria = vendas_mes / len(vendas_agosto)

# Cálculos de Vendas Semanais
semana1 = {}
semana2 = {}
semana3 = {}
semana4 = {}

for i in range(0, 5):
    semana1[lista_datas[i]] = valores_vendas[i]
    semana2[lista_datas[i + 5]] = valores_vendas[i + 5]
    semana3[lista_datas[i + 10]] = valores_vendas[i + 10]
    semana4[lista_datas[i + 15]] = valores_vendas[i + 15]

total_sem1 = sum(list(semana1.values()))
total_sem2 = sum(list(semana2.values()))
total_sem3 = sum(list(semana3.values()))
total_sem4 = sum(list(semana4.values()))

# Cálculo de Máximo e Mínimo
valor_max = max(valores_vendas)
dia_maior_venda_index = valores_vendas.index(valor_max)
dia_maior_venda = lista_datas[dia_maior_venda_index]

valor_min = min(valores_vendas)
dia_menor_venda_index = valores_vendas.index(valor_min)
dia_menor_venda = lista_datas[dia_menor_venda_index]

# 
# RELATÓRIO DE VENDAS DE AGOSTO
#

print("\n" + "="*40)
print("     RELATÓRIO DE VENDAS DE AGOSTO")
print("="*40 + "\n")

# Resumo Mensal
print("## Resumo Mensal")
print("----------------------------------------")
print(f"**Total de Vendas no Mês:** {vendas_mes:,.2f}")
print(f"**Média Diária de Vendas:** {media_diaria:,.2f}")
print(f"**Total de Dias de Venda:** {len(vendas_agosto)}")
print("----------------------------------------\n")

# Desempenho Semanal
print("## Desempenho Semanal")
print("----------------------------------------")
print(f"Vendas da **Semana 1** ({list(semana1.keys())[0]} a {list(semana1.keys())[-1]}): {total_sem1:,.2f}")
print(f"Vendas da **Semana 2** ({list(semana2.keys())[0]} a {list(semana2.keys())[-1]}): {total_sem2:,.2f}")
print(f"Vendas da **Semana 3** ({list(semana3.keys())[0]} a {list(semana3.keys())[-1]}): {total_sem3:,.2f}")
print(f"Vendas da **Semana 4** ({list(semana4.keys())[0]} a {list(semana4.keys())[-1]}): {total_sem4:,.2f}")
print("----------------------------------------\n")

# Destaques (Máximo e Mínimo) 
print("## Destaques do Mês")
print("----------------------------------------")
print(f"**Maior Venda:** {valor_max:,.2f}")
print(f"   -Feita no dia: **{dia_maior_venda}**")
print(f"**Menor Venda:** {valor_min:,.2f}")
print(f"   Feita no dia: **{dia_menor_venda}**")
print("----------------------------------------\n")

```
